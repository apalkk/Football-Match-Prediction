```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("Fifa_world_cup_matches.csv")
df.head(20)
```

```python
df.drop(["team1","team2","date","hour","category","conceded team1","conceded team2","number of goals team1","number of goals team2"], axis = 1, inplace=True)
df['possession team1'] = df['possession team1'].map(lambda x: x.lstrip('+-').rstrip('%'))
df['possession team2'] = df['possession team2'].map(lambda x: x.lstrip('+-').rstrip('%'))
df['possession in contest'] = df['possession in contest'].map(lambda x: x.lstrip('+-').rstrip('%'))
df
```

```python
train_set,test_set = train_test_split(df, test_size=0.2, random_state = 13)
```

```python
x_train = train_set.drop(['goalDiff'], axis = 1)
y_train = train_set['goalDiff'].copy()
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
train_preds = lr.predict(x_train)
mse = mean_squared_error(train_preds,y_train)
print(np.sqrt(mse))
```
