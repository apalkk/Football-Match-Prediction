```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
path='Fifa_world_cup_matches.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.drop(["team1","team2","date","hour","category"], axis = 1, inplace=True)
```

```python
df['possession team1'] = df['possession team1'].map(lambda x: x.lstrip('+-').rstrip('%'))
df['possession team2'] = df['possession team2'].map(lambda x: x.lstrip('+-').rstrip('%'))
df['possession in contest'] = df['possession in contest'].map(lambda x: x.lstrip('+-').rstrip('%'))
df
```

```python
train_set, test_set = train_test_split(df, test_size=0.2, random_state = 42)
train_set.head()
```

```python
y_train = train_set["number of goals team1"].copy()
x_train = train_set.drop("number of goals team1",axis=1)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
train_preds = lr.predict(x_train)
mse = mean_squared_error(train_preds,y_train)
print(np.sqrt(mse))
```

```python
y_test = test_set["number of goals team1"].copy()
x_test = test_set.drop("number of goals team1",axis=1)
```

```python
lr.fit(x_test,y_test)
test_preds = lr.predict(x_test)
mse = mean_squared_error(y_test,test_preds)
print(np.sqrt(mse))
```
