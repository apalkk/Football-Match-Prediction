```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("Fifa_world_cup_matches.csv")
df.head(20)
```

```python
df['possession team1'] = df['possession team1'].map(lambda x: x.lstrip('+-').rstrip('%'))
df['possession team2'] = df['possession team2'].map(lambda x: x.lstrip('+-').rstrip('%'))
df['possession in contest'] = df['possession in contest'].map(lambda x: x.lstrip('+-').rstrip('%'))
df = df[["possession team1","possession team2","possession in contest","goal preventions team1","goal preventions team2","forced turnovers team1","forced turnovers team2","goalDiff"]].copy(deep=True)
df
```

```python
train_set,test_set = train_test_split(df, test_size=0.7, random_state = 13)
```

```python
x_train = train_set.drop(['goalDiff'], axis = 1)
y_train = train_set['goalDiff'].copy()
```

```python
lr = LinearRegression(fit_intercept=False,n_jobs = 1)
lr.fit(x_train,y_train)
```

```python
train_preds = lr.predict(x_train)
mse = mean_squared_error(train_preds,y_train)
print(np.sqrt(mse))
```

```python
x_test = test_set.drop(['goalDiff'], axis = 1)
y_test = test_set['goalDiff'].copy()
```

```python
train_preds = lr.predict(x_test)
mse = mean_squared_error(train_preds,y_test)
print(np.sqrt(mse))
```
